```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
split_1_ranks = ["Iron", "Bronze", "Silver", "Gold", "Platinum", "Diamond", "Apex"]
split_2_ranks = ["Iron", "Bronze", "Silver", "Gold", "Platinum", "Emerald", "Diamond", "Apex"]

split_1 = np.array([
    6.3, 25, 31, 23, 11.05, 3, 0.65
])

split_2 = np.array([
    4, 16, 20, 20, 20, 17, 2.5, 0.5
])

split_1_colors = ["dimgray", "peru", "silver", "gold", "olivedrab", "cyan", "red"]
split_2_colors = ["dimgray", "peru", "silver", "gold", "olivedrab", "lime", "cyan", "red"]

```

```python
plt.figure(figsize=(10, 5))
plt.bar(split_1_ranks, split_1, color=split_1_colors)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(split_2_ranks, split_2, color=split_2_colors)
```

```python
import pandas as pd

# these are hand calculated:
new_split = [[(4, 0)],
             [(2.5, 0), (13.5, 1)],
             [(11.5, 1), (8.5, 2)],
             [(20, 2)],
             [(2.5, 2), (17.5, 3)],
             [(5.5, 3), (11.05, 4), (0.45, 5)],
             [(2.5, 5)],
             [(0.05, 5), (0.65, 6)]
             ]

pre_df = []

for x, i in enumerate(new_split):
    temp = [split_2_ranks[x], 0, 0, 0, 0, 0, 0, 0]
    for j, elo in i:
        temp[elo + 1] = j
    pre_df.append(temp)
```

```python
df = pd.DataFrame(pre_df, columns=["Elo", "Iron", "Bronze", "Silver", "Gold", "Platinum", "Diamond", "Apex"])
print(df)

df.plot(x="Elo", kind="bar", stacked=True, color=split_1_colors)
```
